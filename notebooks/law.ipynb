```python
with open(law_actual_path, 'r', encoding='cp1251') as file:
    law_actual = file.read()
with open(law_old_path, 'r', encoding='cp1251') as file:
    law_old = file.read()

# print(law_actual)
```

```python
import re
import pandas as pd

# Исходный текст
text = law_actual

# Функция для извлечения структуры документа и подпунктов
def extract_structure(text):
    pattern = (r'(?P<Section>Раздел III.*?)(?=Раздел IV|$)|'
               r'(?P<Subsection>Подраздел \d.*?)(?=Подраздел \d|$)|'
               r'(?P<Chapter>Глава \d+.*?)(?=Глава \d+|$)|'
               r'(?P<Paragraph>§ \d+.*?)(?=§ \d+|$)|'
               r'(?P<Article>Статья \d+(\.\d+)?\..*?)(?=Статья \d+(\.\d+)?\.|$)')
    
    items = []
    for match in re.finditer(pattern, text, re.DOTALL):
        section = match.group('Section')
        subsection = match.group('Subsection')
        chapter = match.group('Chapter')
        paragraph = match.group('Paragraph')
        article = match.group('Article')
        
        if article:
            # Извлечение пунктов и подпунктов внутри статьи
            points_pattern = (r'(?P<Point>\d+\.\s.*?)(?=\d+\.\s|$)|'
                              r'(?P<Subpoint>\d+\.\d+\.\s.*?)(?=\d+\.\d+\.\s|$)')
            for point_match in re.finditer(points_pattern, article, re.DOTALL):
                point = point_match.group('Point')
                subpoint = point_match.group('Subpoint')
                items.append({
                    "Section": section,
                    "Subsection": subsection,
                    "Chapter": chapter,
                    "Paragraph": paragraph,
                    "Article": article,
                    "Point": point,
                    "Subpoint": subpoint
                })
        else:
            items.append({
                "Section": section,
                "Subsection": subsection,
                "Chapter": chapter,
                "Paragraph": paragraph,
                "Article": article,
                "Point": None,
                "Subpoint": None
            })
    
    return pd.DataFrame(items)

# Создание DataFrame из исходного текста
df = extract_structure(text)
df
```
