```python
article_pattern = re.compile(r'^(Статья \d+\..*?)(?=\d+\.)', re.DOTALL)

# Список для хранения разделов и пунктов
result_list = []

# Итерация по списку текстов разделов и статей
for item in temp_:
    # Если строка начинается с "Статья"
    if article_pattern.search(item):
        # Разбиваем статью на номер и название
        article_parts = article_pattern.findall(item)
        # Добавляем номер и название статьи в список результатов
        result_list.extend(article_parts)
        # Разбиваем оставшуюся часть статьи на пункты
        article_body = re.findall(r'\n\s+\d+\..*?(?=\n\s+\d+\.|\Z)', item, re.DOTALL)
        # Добавляем каждый пункт в список результатов
        result_list.extend(article_body[1:])
    else:
        # Если строка не начинается с "Статья", добавляем её как есть в список результатов
        result_list.append(item)
```
